count  function  > show the counts
splete function >  create a list 


```python
sent_1 = "depi learning depi is fun"
print(sent_1.count("depi"))  # Output: 2
print(sent_1.split())        # Output: ['depi', 'learning', 'depi', 'is', 'fun']
```
