```python
list_1 = ["data", "engineer", "is", "good", [1, 2, 3, 4,True,["ML", "DL"]]]

#find DL in the nested list
DL_IN = list_1[4][5][1]
print(DL_IN)  # Output: DL
#make string to list
chang_tex_to_list = list("DataEngineer")
print(chang_tex_to_list)
#append to list at the end
list_1.append("Welcome")
list_1.append(2025)
print(list_1)
#list insert at specific index
list_1.insert(2, "Welcome to")
print(list_1)
#extend list with another list like append but for multiple values at once 
list_2 = ["Element1", "Element2"]
# extend accepts a single iterable; combine the lists into one iterable
list_1.extend(["New Element1", "New Element2"] + list_2)
list_1.extend(list_2)
list_2.extend("mo")
print(list_2)
print(list_1)

# dunction delete list element by value
list_1.remove("Welcome") # Removes the first occurrence of "Welcome" specified by value , output of remove is None not print the modified list  
print(list_1)
del list_1[3]  # Deletes the element at index 3
print(list_1)
# pop function delete list element by index and return the deleted value
popped_value = list_1.pop(4)  # Removes and returns the element at index 4 , default is -1 last element 
print("Popped Value:", popped_value)
print(list_1)

tuple_1 = ("data", "engineer", "is", "good", (1, 2, 3, 4,False,("ML", "DL")))
#find DL in the nested tuple
DL_IN_TUPLE = tuple_1[4][5][1]
print(DL_IN_TUPLE)  # Output: DL
```
