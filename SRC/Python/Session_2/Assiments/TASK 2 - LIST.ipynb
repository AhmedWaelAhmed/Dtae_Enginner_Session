### <span style="color:orange"><u>**Exercise 1: Creat List**<u></span>

- Ask the user to enter 5 numbers

- Store them in a list

- Print the first and last element

- Hint:
  - Use input().split() and list indexing

```python
list_1=[]
for i in range(5):
    enter_number= int(input("enter 5 numbers:  "))
    list_1.append(enter_number)

print("list is :",list_1)
print("first element ",list_1[0],"last element ",list_1[4])

# another method using spilt function
list_3 = input("enter 5 numbers with space:  ").split()
print("list is :",list_3)
print("first element ",list_3[0],"last element ",list_3[4])
```

### <span style="color:orange"><u>**Exercise 5: Sorting**<u></span>

- Input numbers
- Sort ascending and descending


```python
list_nonso=[]
for i in range(5):
    numers= int(input("enter numbers need to sort"))
    list_nonso.append(numers)
print("the original list",list_nonso)
list_nonso.sort()

print("orderd list from small to big",list_nonso)
list_nonso.reverse()
print("orderd list from big to small",list_nonso)

```
