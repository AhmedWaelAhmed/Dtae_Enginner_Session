```python
tuple1 = (1,2,3)
print(type(tuple1))
# can not change tuple values tuple1[0] = 10
# we can make it in a list and change it
list1 = list(tuple1)
list1[0] = 10
print(list1)
```

```python
#set is not ordered and not indexed, no duplicate values
set_1={1,2,3,4,5,5,6,7,8}
print(set_1)
print(type(set_1))
Set_2={5,4,3,2,1}
print(set_1.union(Set_2))
print(set_1.intersection(Set_2))
print(set_1.difference(Set_2))
```

```python
#dictionary key-value pairs
dict_1={"name":{"Ahmed","ALI"},"age":{22,24,24},"city":"sohag"}
print(dict_1["name"])
dict_1["age"]=23 #update value 
print(dict_1.keys()) #get all keys
print(dict_1.values()) #get all values
print(dict_1.items()) #get all items

print(dict_1)
```

```python
#CHANGE TUPLE TO DI
tuple2 = (("name","Ahmed"),("age",{22,45,66}),("city","sohag"))
dict_2 = dict(tuple2)
print(dict_2)
```

```python
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))       
sum = number1 + number2
div = number1 / number2
diff = number1 - number2
mult = number1 * number2
print(f"The sum of {number1} and {number2} is {sum}")
print(f"The difference of {number1} and {number2} is {diff}")
print(f"The product of {number1} and {number2} is {mult}")
print(f"The division of {number1} and {number2} is {div}")
```

```python
#floor division
a = 7
b = 3
result = a // b
# 7 // 3 = 2.666666 = 2
print(f"The floor division of {a} by {b} is {result}")  
#modulus
mod_result = a % b
print(f"The modulus of {a} by {b} is {mod_result}")
#3 * 2 = 6 + 1 = 7
```
