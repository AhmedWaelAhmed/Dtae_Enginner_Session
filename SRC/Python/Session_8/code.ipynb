```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary         # Public (No underscore)
        
    def set_salary(self, obj, new_sal):
        obj.salary = new_sal         # accessing public attribute

class Manager:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary         # Public (No underscore)

    def get_salary(self):
        return self.salary           # accessing public attribute

# --- Testing the code ---
e1 = Employee("Depi", 100)
m1 = Manager("Ahmed", 120)

e1.set_salary(m1, 500)

print(m1.salary)       # Output: 500 (Accessing directly works fine)
print(m1.get_salary()) # Output: 500
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self._salary = salary        # Added underscore (Protected)
        
    def set_salary(self, obj, new_sal):
        obj._salary = new_sal        # Added underscore to access the protected variable

class Manager:
    def __init__(self, name, salary):
        self.name = name
        self._salary = salary        # Added underscore (Protected)

    def get_salary(self):
        return self._salary          # Added underscore to return the protected variable

# --- Testing the code ---
e1 = Employee("Depi", 100)
m1 = Manager("Ahmed", 120)

# Now this updates the protected variable '_salary' inside m1
e1.set_salary(m1, 500)

print(m1.get_salary())  # Output: 500
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary  # üîí (Private) ŸÖŸÇŸÅŸàŸÑ ÿπŸÑŸäŸá
    
    # ÿØÿßŸÑÿ© ÿπÿ¥ÿßŸÜ "ŸÜÿ¥ŸàŸÅ" ÿßŸÑŸÖÿ±ÿ™ÿ® (Getter)
    def get_salary(self):
        return self.__salary

    # ÿØÿßŸÑÿ© ÿπÿ¥ÿßŸÜ "ŸÜÿπÿØŸÑ" ÿßŸÑŸÖÿ±ÿ™ÿ® (Setter)
    def set_salary(self, new_salary):
        # ÿßŸÑŸÖŸäÿ≤ÿ© ŸáŸÜÿß ÿ•ŸÜŸÜÿß ŸÖŸÖŸÉŸÜ ŸÜÿ≠ÿ∑ ÿ¥ÿ±Ÿàÿ∑ (Validation)
        if new_salary > 0:
            self.__salary = new_salary
            print(f"Salary updated to: {self.__salary}")
        else:
            print("Error: Salary cannot be negative!")

# --- ŸÜÿ¨ÿ±ÿ® ÿßŸÑŸÉŸàÿØ ---
emp = Employee("Depi", 1000)

# 1. ŸÑŸà ÿ≠ÿßŸàŸÑÿ™ ÿ™ŸàÿµŸÑ ŸÖÿ®ÿßÿ¥ÿ± ŸáŸäŸÅÿ¥ŸÑ
# print(emp.__salary)  # ‚ùå Error

# 2. ŸÜÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑÿ∑ÿ±ŸäŸÇÿ© ÿßŸÑÿ±ÿ≥ŸÖŸäÿ© (Getter)
print(emp.get_salary())  # ‚úÖ ŸáŸäÿ∑ÿ®ÿπ: 1000

# 3. ŸÜÿπÿØŸÑ ÿßŸÑŸÖÿ±ÿ™ÿ® ÿ®ÿßŸÑÿ∑ÿ±ŸäŸÇÿ© ÿßŸÑÿ±ÿ≥ŸÖŸäÿ© (Setter)
emp.set_salary(5000)     # ‚úÖ ŸáŸäÿ∑ÿ®ÿπ: Salary updated to: 5000

# 4. ŸÜÿ¨ÿ±ÿ® ŸÜÿ∂ÿ≠ŸÉ ÿπŸÑŸäŸá ŸàŸÜÿ≠ÿ∑ ÿ±ŸÇŸÖ ÿ®ÿßŸÑÿ≥ÿßŸÑÿ® (ŸÖŸäÿ≤ÿ© ÿßŸÑÿ≠ŸÖÿßŸäÿ©)
emp.set_salary(-200)     # ‚ùå ŸáŸäÿ∑ÿ®ÿπ: Error: Salary cannot be negative!
```

```python
from prettytable import PrettyTable

products = [{"name":"Water", "price":80.0, "quantity":1500},
            {"name":"Soda", "price":130.0, "quantity":1500},
            {"name":"Chips", "price":75.0, "quantity":1500},
            {"name":"Eggs", "price":200.0, "quantity":1500},
           ]

table = PrettyTable()

# 1. ŸáŸÜÿß ÿ®ŸÜÿ≠ÿØÿØ ÿ£ÿ≥ŸÖÿßÿ° ÿßŸÑÿπŸàÿßŸÖŸäÿØ (ŸÑÿßÿ≤ŸÖ ÿ™ŸÉŸàŸÜ ŸÑŸäÿ≥ÿ™)
table.field_names = ["Product Name", "Price", "Quantity"]

# 2. ŸàÿØŸÑŸàŸÇÿ™Ÿä ŸÜŸÑŸÅ ÿπŸÑŸâ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸàŸÜÿ∂ŸäŸÅŸáÿß ÿµŸÅ ÿ®ÿµŸÅ
for item in products:
    table.add_row([item["name"], f"{item["price"]:.2f}", item["quantity"]])

print(table)
```

```python
#session 8
# You need a product list for this to work. I added a dummy one here:
products = [
    {'name': 'Bread', 'price': 45.00, 'quantity': 1500},
    {'name': 'Eggs', 'price': 200.00, 'quantity': 1500}
]

total_bill = 0

while True:
    product_name = input("\nEnter product name  (or 'exit) to finish").strip().lower()
    if product_name == 'exit':
        break
    
    # Search for product
    product = next((p for p in products if p['name'].lower() == product_name), None)
    
    if product is None:
        print("product not found. please try again")
        continue

    try:
        quantity_required = float(input(f"how many {product['name']} do you want?"))
        
        if quantity_required <= 0:
            print("Quantity must be greater than zero")
            continue
            
        if quantity_required > product['quantity']:
            print("Insufficient quantity!")
            continue

        # --- Calculations ---
        discount_precentage = 0.05
        price_before_disc = quantity_required * product['price']
        
        # FIX: In your screenshot, you multiplied by 'discount_amount' which 
        # doesn't exist yet. I changed it to 'discount_precentage'.
        discount_amount = price_before_disc * discount_precentage 
        
        final_price = price_before_disc - discount_amount

        # --- Updates ---
        product['quantity'] -= quantity_required
        total_bill += final_price
        
        print(f"Success! price for {product['name']}: $ {final_price:.2f}")
        
        another = input("something else? (y/n)").lower()
        if another != 'y':
            break
            
    except ValueError:
        print("Invalid input! pls enter a number for QTY")

# --- Final Output ---
print(f"\n total amount to pay : ${total_bill:.2f}")
print("thank you !!!!!!!!!!!!!!!!!!!")
# task to make it in the program
```
