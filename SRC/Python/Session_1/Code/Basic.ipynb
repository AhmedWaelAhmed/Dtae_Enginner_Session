Make all letters uppercase
The .upper() method converts all characters in the string to uppercase.

```python
name = "welcome into data engineer course"
print(name.upper()) 
# Output: WELCOME INTO DATA ENGINEER COURSE
```

Make the first letter of every word uppercase
The .title() method converts the first character of each word to uppercase.

```python
name = "welcome into data engineer course"
print(name.title())
# Output: Welcome Into Data Engineer Course
```

Make only the first letter of the sentence uppercase
The .capitalize() method converts the first character of the string to uppercase and the rest to lowercase.

```python
name = "welcome into data engineer course"
print(name.capitalize())
# Output: Welcome into data engineer course
```

Make all letters lowercase
The .lower() method converts all characters in the string to lowercase.

```python
name = "WELCOME INTO DATA ENGINEER COURSE"
print(name.lower())
# Output: welcome into data engineer course
```
