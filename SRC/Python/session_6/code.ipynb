```python
# we want to build a factorial app takes input number 
def factorila_1(n1:int):
    if n1 ==0 :
        return 1
    elif n1==1:
        return 1
    return n1*factorila_1(n1-1) 
```

```python
def even(n:int):
    l2=[]

    for i in range(n):
        if i%2==0:
            l2.append(i)
    return l2
```

المشكلة / The Problem
بالعربية: أوجد الكلمات الفريدة واحسب تكرار كل كلمة من قائمة نصوص معطاة.

المطلوب: كتابة برنامج يأخذ قائمة من الكلمات، ثم يقوم بإنشاء قاموس (Dictionary) يحتوي على كل كلمة وعدد المرات التي ظهرت فيها داخل القائمة.

In English: Find all unique words and count the frequency of occurrence from a given list of strings.

Task: Write a program that takes a list of strings and creates a dictionary where each key is a unique word from the list and its value is the number of times it appears.

مثال / Example
Input (المُدخلات): ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]

Output (المُخرجات المتوقعة): {'Ali': 4, 'Welcome': 1, 'No': 3, 'Hi': 2}

```python
def ge(words_list):
    unique_words = set(words_list)
    
    counts = {}
    
    for word in unique_words:
        count_value = words_list.count(word)
        
        counts[word] = count_value
        
    return counts

words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]
final_result = ge(words)
print(final_result)
```

```python
import random

responses = {
    "hello": ["Hello!", "Hi there!", "Greetings!"],
    "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
    "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
    "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
}

def get_response(user_input):
    for key in responses:
        if key in user_input:
            return random.choice(responses[key])
    return random.choice(responses["default"])

def chatbot():
    print("Chatbot: Hi! How can i assist you today?")
    while True:
        user_input = input("User: ").lower()
        response = get_response(user_input)
        print("Chatbot: ", response)
        
        if user_input == "goodbye":
            break

# I added this line so the chatbot starts when you run the code
chatbot()
```
