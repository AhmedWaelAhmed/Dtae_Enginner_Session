```python
import numpy as np

# --- 1. Data Setup ---
# data_x: Your original dataset
data_x = np.array([10, 20, 20, 30, 40, 50, 60, 70, 80, 90])

# data_y: A second dataset required to calculate Covariance & Correlation
# I created this array to have a strong positive relationship with data_x
data_y = np.array([12, 22, 25, 35, 42, 55, 62, 75, 82, 95])

# --- 2. Descriptive Statistics (Single Variable) ---
mean_val = np.mean(data_x)
median_val = np.median(data_x)

# Mode: numpy has no direct mode function. 
# bincount() counts occurrences, argmax() finds the index of the highest count.
mode_val = np.bincount(data_x).argmax() 

# Variance & Standard Deviation
# ddof=0 is Population Variance (divide by N)
# ddof=1 is Sample Variance (divide by N-1) - Standard in research/statistics
variance = np.var(data_x, ddof=1)
std_dev = np.std(data_x, ddof=1)

min_val = np.min(data_x)
max_val = np.max(data_x)

# Percentiles: 25th (Q1), 50th (Median), 75th (Q3)
percentiles = np.percentile(data_x, [25, 50, 75])

# --- 3. Relationship Statistics (Two Variables) ---

# Covariance
# Returns a 2x2 matrix: [[var(x), cov(x,y)], [cov(y,x), var(y)]]
# We select [0,1] to get the covariance between x and y
cov_matrix = np.cov(data_x, data_y, ddof=1)
covariance_val = cov_matrix[0, 1]

# Correlation
# Returns a 2x2 matrix of correlation coefficients (r)
# r is always between -1 and 1
corr_matrix = np.corrcoef(data_x, data_y)
correlation_val = corr_matrix[0, 1]

# --- 4. Comparison (Fixing your NameError) ---
# np.isclose checks if two arrays are similar within a tolerance (atol)
# This is useful for cleaning data or comparing float values
are_close = np.isclose(data_x, data_y, atol=5) # Checks if they are within 5 units

# --- 5. Printing Results ---
print("--- DESCRIPTIVE STATISTICS (Data X) ---")
print(f"Mean:       {mean_val}")
print(f"Median:     {median_val}")
print(f"Mode:       {mode_val}")
print(f"Variance:   {variance:.2f}")
print(f"Std Dev:    {std_dev:.2f}")
print(f"Min/Max:    {min_val} / {max_val}")
print(f"Percentiles:{percentiles}")

print("\n--- RELATIONSHIP STATISTICS (X vs Y) ---")
print(f"Covariance:  {covariance_val:.2f}")
print(f"Correlation: {correlation_val:.4f}")

print("\n--- COMPARISON (isclose) ---")
print(f"Are X and Y within 5 units of each other?\n{are_close}")
```

```python
import numpy as np

# 1. DEFINE the variables first
v1 = np.array([10, 20, 20, 30, 40, 50, 60, 70, 80, 90])
v2 = np.array([15, 25, 10, 35, 45, 50, 70, 65, 85, 95])

# 2. NOW you can compare them
# This checks: Is v1 close to v2 (within a difference of 1)?
result = np.isclose(v1, v2, atol=1)

print("v1:", v1)
print("v2:", v2)
print("Are they close?", result)
```
