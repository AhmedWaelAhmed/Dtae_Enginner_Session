```python
try:
    x = 10
    
    # --- Instructions for Testing (تعليمات التجربة) ---
    # 1. To trigger ZeroDivisionError: Enter 0
    # 2. To trigger ValueError: Enter text like "ahmed" or float symbols like "5.5"
    # 3. To trigger generic Exception: Press Ctrl+D (on Linux/Mac) or Ctrl+Z (on Windows)
    y = int(input("Enter a number to divide 10 by: "))
    
    z = x / y
    print("The result is:", z)

except ZeroDivisionError:
    # This block runs if Input is: 0
    print("Error: Cannot divide by zero.")
except ValueError:
    # This block runs if Input is: "text" or "symbols"
    print("Error: Invalid input. Please enter a valid integer.")

except Exception as e:
    # This block runs if Input is: EOF (Ctrl+D / Ctrl+Z)
    print("An unexpected error occurred:", str(e))

finally:
    # This runs ALWAYS (Success or Fail)
    print("--- End of Calculation ---")
```

```python
try:
    x = 10
    # هنا لو أدخلت 0 أو نص خطأ، كله سيروح للـ Exception العام
    y = int(input("Enter a number to divide 10 by: "))
    z = x / y
    print("The result is:", z)

except Exception as e:
    # هذا السطر سيمسك "أي" خطأ مهما كان نوعه
    print("Something went wrong:", str(e))

finally:
    print("--- End of Calculation ---")
```
