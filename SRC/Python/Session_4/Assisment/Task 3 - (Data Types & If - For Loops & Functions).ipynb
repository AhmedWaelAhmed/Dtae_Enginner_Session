### <span style="color:orange"><u>**18. Sort a tuple by its float element**<u></span>

```python
items = (("item1", 10.5), ("item2", 5.7), ("item3", 8.9))

# TODO: sort by the float value
# your code here
def get_price(item):
    return item[1] # index of the float value
    

# Pass the function name (without parentheses) to the key argument
sorted_items = sorted(items, key=get_price)

print(sorted_items)
# Hint: Use sorted() with a key function.

```
