```python
# --- Helper Functions (Implied but not shown in image) ---
def sum(x, y):
    return x + y

def sub(x, y):
    return x - y

def mul(x, y):
    return x * y

def div(x, y):
    if y == 0:
        return "Error! Division by zero"
    return x / y

# --- Main Function (Shown in image) ---
def main():
    while True:
        print("\n Calculator")
        print("1. sum")
        print("2. sub")
        print("3. mul")
        print("4. div")
        # Note: Option 5 is hidden in the menu but works in the code
        
        choice = input("Enter your choice: ")

        if choice == '5':
            print("calc stopped")
            break

        # We convert to float to handle decimals
        a = float(input("enter first number: "))
        b = float(input("enter second number: "))

        if choice == '1':
            print("Result: ", sum(a, b))
        elif choice == '2':
            print("Result: ", sub(a, b))
        elif choice == '3':
            print("Result: ", mul(a, b))
        elif choice == '4':
            print("Result: ", div(a, b))
        else:
            print("invalid choice")

if __name__ == "__main__":
    
    main()
```
